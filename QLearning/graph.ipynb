```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the data
data = pd.read_csv(r'C:\Users\tyler\OneDrive\Desktop\GitHub Repositories\RFP3\QLearning\training_testing_statistics.csv')

# Display the first few rows of the DataFrame to ensure it's loaded correctly
print(data.head())
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Set the aesthetic style of the plots
sns.set(style="whitegrid")

# Create a figure to hold the subplots
fig, axes = plt.subplots(3, 1, figsize=(10, 15))

# Plot histogram and KDE for Time to Convergence
sns.histplot(data['Time to Convergence (seconds)'], kde=True, ax=axes[0], color='blue', bins=30)
axes[0].set_title('Distribution of Time to Convergence')
axes[0].set_xlabel('Time to Convergence (seconds)')
axes[0].set_ylabel('Frequency')

# Plot histogram and KDE for Episodes to Convergence
sns.histplot(data['Episodes to Convergence'], kde=True, ax=axes[1], color='green', bins=30)
axes[1].set_title('Distribution of Episodes to Convergence')
axes[1].set_xlabel('Episodes to Convergence')
axes[1].set_ylabel('Frequency')

# Plot histogram and KDE for Average Steps during Testing
sns.histplot(data['Average Steps (Testing)'], kde=True, ax=axes[2], color='red', bins=30)
axes[2].set_title('Distribution of Average Steps during Testing')
axes[2].set_xlabel('Average Steps')
axes[2].set_ylabel('Frequency')
axes[2].set_xlim(0, 12)  # Setting x-axis limits to focus around the peak value

# Adjust layout and display the plots
plt.tight_layout()
plt.show()
```

```python
# Set the number of testing episodes per training session and the number of training sessions
num_test_episodes_per_session = 10000  # replace with the number you used for testing
num_training_sessions = 1000  # replace with the number you used for training sessions

# Sum of 'Successes (Testing)'
total_successes = data['Successes (Testing)'].sum()

# Total number of episodes tested
total_episodes_tested = num_test_episodes_per_session * num_training_sessions

# Calculate success percentage
success_percentage = (total_successes / total_episodes_tested) * 100

print(f"Total successes: {total_successes}")
print(f"Total episodes tested: {total_episodes_tested}")
print(f"Success percentage: {success_percentage}%")
```
